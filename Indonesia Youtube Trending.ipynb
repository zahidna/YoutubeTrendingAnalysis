```python
#Drop some useless data and show table after some useless data is dropped
df = df.drop(['video_id', 'channelId', 'categoryId', 'trending_date', 'tags', 'thumbnail_link', 'comments_disabled', 'ratings_disabled', 'description'], axis=1)
df.head(10)
```

```python
#View some basic statistical details
df.describe()
```

```python
#Detailed average views, likes, dislikes, and comment
avgviews= df["view_count"].mean()
avglikes= df["likes"].mean()
avgdislikes= df["dislikes"].mean()
avgcomments= df["comment_count"].mean()

print("Average Views   : ", avgviews)
print("Average Likes   : ", avglikes)
print("Average Dislikes: ", avgdislikes)
print("Average Comments: ", avgcomments)
```

```python
df.head(10)
```

```python
#Trending videos with views above average (42 videos)
df.loc[(df.view_count >= avgviews)]
```

```python
#Trending videos with likes above average (34 videos)
df.loc[(df.likes >= avglikes)]
```

```python
#Trending videos with dislikes above average (31 videos)
df.loc[(df.dislikes >= avgdislikes)]
```

```python
#Trending videos with comments above average (24 videos)
df.loc[(df.comment_count >= avgcomments)]
```

```python
#Top 10 most viewed video
MostViewed= df.sort_values(by='view_count', ascending=False)
MostViewed.iloc[0:10]
```

```python
#Top 10 most liked video
MostLikes= df.sort_values(by='likes', ascending=False)
MostLikes.iloc[0:10]
```

```python
#Top 10 most disliked video
MostDislikes= df.sort_values(by='dislikes', ascending=False)
MostDislikes.iloc[0:10]
```

```python
#Top 10 most commented video
MostCommented= df.sort_values(by='comment_count', ascending=False)
MostCommented.iloc[0:10]
```
